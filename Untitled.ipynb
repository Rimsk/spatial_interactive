```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# 1. Extract gene expression and coordinates
gene = 'BDP1'
gene_idx = adata.var_names.get_loc(gene)
gene_expr = adata.X[:, gene_idx].toarray().flatten()

coords = adata.obsm['spatial']
x, y = coords[:, 0], coords[:, 1]

# 2. Create a grid over the spatial domain
xi = np.linspace(np.min(x), np.max(x), 300)
yi = np.linspace(np.min(y), np.max(y), 300)
xi, yi = np.meshgrid(xi, yi)

# 3. Interpolate expression values onto grid
zi = griddata((x, y), gene_expr, (xi, yi), method='cubic')  # or 'linear'

# 4. Plot heatmap
plt.figure(figsize=(8, 6))
heat = plt.imshow(zi, origin='lower', extent=(x.min(), x.max(), y.min(), y.max()),
                  cmap='viridis', aspect='auto')
plt.colorbar(heat, label=f'{gene} Expression')
plt.title(f'Spatial Heatmap of {gene}')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  # Optional
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
from IPython.display import display, clear_output

# --- Define your T-cell marker genes
tcell_genes = ['CD3D', 'CD3E', 'CD4', 'CD8A', 'CD8B']

# --- Create widgets
gene1_input = widgets.Text(value='BDP1', description='Gene:')
gene2_dropdown = widgets.Dropdown(options=sorted(tcell_genes), value='CD3D', description='T cell:')

# --- Create output area to control plotting
plot_output = widgets.Output()

# --- Define plotting function
def plot_gene(gene):
    with plot_output:
        clear_output(wait=True)
        try:
            gene_idx = adata.var_names.get_loc(gene)
            gene_expr = adata.X[:, gene_idx].toarray().flatten()
            coords = adata.obsm['spatial']
            x, y = coords[:, 0], coords[:, 1]

            # Masking for expression
            expressed = gene_expr > 0
            not_expressed = ~expressed

            fig, ax = plt.subplots(figsize=(8, 6), facecolor='lightgrey')

            # Non-expressed: grey
            ax.scatter(x[not_expressed], y[not_expressed], c='lightgrey', s=35)

            # Expressed: colored
            sc = ax.scatter(x[expressed], y[expressed], c=gene_expr[expressed], cmap='magma_r', s=35)
            plt.colorbar(sc, ax=ax, label=f'{gene} Expression')

            ax.set_xlabel('Tissue X Coordinate')
            ax.set_ylabel('Tissue Y Coordinate')
            ax.set_title(f'Spatial Heatmap of {gene}')
            ax.invert_yaxis()
            plt.tight_layout()
            plt.show()

        except KeyError:
            print(f"Gene '{gene}' not found in the dataset.")

# --- Event handlers for widgets
def on_gene1_submit(change):
    plot_gene(change['new'])

def on_gene2_change(change):
    plot_gene(change['new'])

# --- Attach observers
gene1_input.observe(on_gene1_submit, names='value')
gene2_dropdown.observe(on_gene2_change, names='value')

# --- Display
display(widgets.HBox([gene1_input, gene2_dropdown]))
display(plot_output)

# --- Initial plot
plot_gene(gene1_input.value)
```

```python
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import numpy as np

# --- Define genes
tcell_genes = ['CD3D', 'CD3E', 'CD4', 'CD8A', 'CD8B', 'BDP1']

# --- Create widgets
gene_input = widgets.Text(value='BDP1', description='Gene:')
gene_dropdown = widgets.Dropdown(options=tcell_genes, value='CD3D', description='T cell:')

output = widgets.Output()

def plot_gene(gene):
    with output:
        clear_output(wait=True)
        try:
            gene_idx = adata.var_names.get_loc(gene)
            gene_expr = adata.X[:, gene_idx].toarray().flatten()
            coords = adata.obsm['spatial']
            x, y = coords[:, 0], coords[:, 1]

            df = pd.DataFrame({
                'x': x,
                'y': y,
                'expression': gene_expr
            })

            # Split into expressed and not expressed
            expressed = df['expression'] > 0
            df_expr = df[expressed]
            df_zero = df[~expressed]

            # Create figure
            fig = go.Figure()

            # Add non-expressed (grey)
            fig.add_trace(go.Scattergl(
                x=df_zero['x'],
                y=df_zero['y'],
                mode='markers',
                marker=dict(color='lightgrey', size=6),
                name='Not Expressed',
                showlegend=False
            ))

            # Add expressed (colored)
            fig.add_trace(go.Scattergl(
                x=df_expr['x'],
                y=df_expr['y'],
                mode='markers',
                marker=dict(
                    color=df_expr['expression'],
                    colorscale='magma_r',
                    size=6,
                    colorbar=dict(title=f'{gene} Expression')
                ),
                name='Expressed',
                showlegend=False
            ))

            fig.update_layout(
                title=f"Spatial Expression of {gene}",
                xaxis_title="Tissue X",
                yaxis_title="Tissue Y",
                yaxis=dict(autorange='reversed'),
                width=700,
                height=600
            )

            fig.show()

        except KeyError:
            print(f"Gene '{gene}' not found in the dataset.")

# --- Event handlers
def on_gene_text_submit(change):
    plot_gene(change['new'])

def on_dropdown_change(change):
    plot_gene(change['new'])

gene_input.observe(on_gene_text_submit, names='value')
gene_dropdown.observe(on_dropdown_change, names='value')

# --- Display
display(widgets.HBox([gene_input, gene_dropdown]))
display(output)

# --- Initial plot
plot_gene(gene_input.value)
```
